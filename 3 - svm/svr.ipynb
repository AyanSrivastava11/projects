# feature encoding - label and OHE

```python
df.columns
```

```python
X= df[['tip', 'sex', 'smoker', 'day', 'time', 'size']]
y = df['total_bill']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le1 = LabelEncoder()
le2 = LabelEncoder()
le3 = LabelEncoder()
```

```python
X_train['sex'] = le1.fit_transform(X_train['sex'])
X_train['smoker'] = le2.fit_transform(X_train['smoker'])
X_train['day'] = le3.fit_transform(X_train['time'])
```

```python
X_train.head()
```

```python
X_test['sex'] = le1.transform(X_test['sex'])
X_test['smoker'] = le2.transform(X_test['smoker'])
X_test['day'] = le3.transform(X_test['time'])
```

```python
X_test.head()
```

```python
## OHE for day

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ct = ColumnTransformer(transformers=[('encoder',OneHotEncoder(drop='first'),['time'])],remainder='passthrough')
```

```python
X_train=ct.fit_transform(X_train)
```

```python
X_train
```

```python
X_test=ct.transform(X_test)
```

```python
X_test
```

```python
from sklearn.svm import SVR
svr = SVR()
```

```python
svr.fit(X_train,y_train)
```

```python
y_pred = svr.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
print('MAE:',mean_absolute_error(y_test,y_pred))
print('MSE:',mean_squared_error(y_test,y_pred))
print('R2 Score:',r2_score(y_test,y_pred))
```

```python
## Hyperparameter Tuning
```

```python
from sklearn.model_selection import GridSearchCV

params = {
    'kernel':['linear','rbf','poly'],
    'C':[0.1,1,10,100],
    'gamma':[1,0.1,0.01,0.001]
}
```

```python
grid = GridSearchCV(SVR(),params,refit=True,verbose=3)

grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid_pred = grid.predict(X_test)
```

```python
print('MAE:',mean_absolute_error(y_test,grid_pred))
print('MSE:',mean_squared_error(y_test,grid_pred))
print('R2 Score:',r2_score(y_test,grid_pred))
```
