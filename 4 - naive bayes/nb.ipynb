```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

```python
X,y = load_iris(return_X_y=True)
```

```python
X
```

```python
y
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.3,random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB
gb = GaussianNB()
```

```python
gb.fit(X_train,y_train)
```

```python
y_pred = gb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
    
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))

print(accuracy_score(y_test, y_pred))
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.columns
```

```python
tips['day'].value_counts()
```

```python
X = tips[['total_bill', 'tip', 'sex', 'smoker', 'time', 'size']]
y = tips['day']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
X_train = X_train.reset_index(drop=True)
X_test = X_test.reset_index(drop=True)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le1 = LabelEncoder()
le2 = LabelEncoder()
le3 = LabelEncoder()
```

```python
X_train['sex']=le1.fit_transform(X_train['sex'])
X_train['smoker']=le2.fit_transform(X_train['smoker'])
X_train['time']=le3.fit_transform(X_train['time'])
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
X_test['sex'] = le1.transform(X_test['sex'].values)
X_test['smoker'] = le2.transform(X_test['smoker'].values)
X_test['time'] = le3.transform(X_test['time'].values)
```

```python
X_test.head()
```

```python
le_target = LabelEncoder()
y_train = le_target.fit_transform(y_train)
y_test = le_target.transform(y_test)
```

```python
gb.fit(X_train,y_train)
```

```python
y_pred_tips = gb.predict(X_test)
```

```python
print(confusion_matrix(y_test, y_pred_tips))
print(classification_report(y_test, y_pred_tips))
print(accuracy_score(y_test, y_pred_tips))
```

```python
from sklearn.naive_bayes import MultinomialNB
mnb = MultinomialNB()
```

```python
mnb.fit(X_train,y_train)
```

```python
y_pred = mnb.predict(X_test)
```
